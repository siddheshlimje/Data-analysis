```python
grouped_counts = df.groupby(['Age Group', 'Gender']).size().unstack(fill_value=0)
```

```python
grouped_counts
```

```python
gender_count=df['Gender'].value_counts()
```

```python
gender_count
```

```python
gender_count.plot(kind='bar',color=['orange'],title='Gender')
plt.show()
```

**from above graph we can say that most of the customers are female**

```python
# Plot grouped bar chart
grouped_counts.plot(kind='bar', color=['skyblue', 'orange'], title='Age Group Distribution by Gender')

# Customize labels and appearance
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=0)  # Ensure labels are horizontal
plt.legend(title='Gender')  # Add legend with title
plt.show()
```

**from the above figure we can say that most of the customers ages between 26-35**

```python
marital_gender_count = pd.crosstab(df['Marital_Status'], df['Gender'])

# Plot the grouped bar chart
marital_gender_count.plot(kind='bar', color=['orange', 'lightgreen'], figsize=(8, 4))
plt.title('Marital Status by Gender')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()
```
